# Helper Functions for Classes:
* The helper functions below will be used in generating data for the mentee/mentor classes.
 * The **random_date** function will be for aiding in generating random timeframes of application submissions.
 * The **percent_true** function will be used in generating boolian outputs from the hypethetical user, or helping other functions.
 * The **random_first_name** function will be purposed for generating either a female, or male name for the hypethetical user.
 * The **random_mentee_language** function will help generate possibly known languages from a mentee.
 * The **random_profile_id** function will be used for generating random profile id's using a hash of the randomly generated email.

```python
def random_date(start_y, start_m, start_d, end_y, end_m, end_d):
  start_date = datetime.date(start_y, start_m, start_d)
  end_date = datetime.date(end_y, end_m, end_d)
  time_between_dates = end_date - start_date
  days_between_dates = time_between_dates.days
  random_number_of_days = random.randrange(days_between_dates)
  random_date = start_date + datetime.timedelta(days=random_number_of_days)
  return random_date

def percent_true(percent: int) -> bool:
  return randint(1, 100) <= percent

def random_first_name():
    return choice(first_names_male if percent_true(75) else first_names_female)

def random_mentee_langauge():
    langauges = ["None", "Python", "HTML", "Java"]
    weights = [75, 15, 5, 5]
    return choices(langauges, weights, k=1)[0]

def random_intention():
    langauges = ["Programming Readiness", "Career Readiness", "Life Coaching",]
    weights = [70, 25, 5]
    return choices(langauges, weights, k=1)[0]

def random_profile_id(email):
  return abs(hash(email))
```

# Mentee Class:

## Random Mentee Sing-up Information
* This generator is used for generating mock data somewhat similar to what a mentee would enter during the on-boarding process.
* The mock data generated from both the mentor/mentee will primarily be used for simulating the pairing process.

```python
class Mentee:
  mentee_intake_id = 0

  def __init__(self):
      self.mentee_intake_id = self.increment_intake_counter()
      self.first_name = random_first_name()
      self.last_name = random.choice(last_names)
      self.email = f"{self.first_name}_{self.last_name}@{random.choice(domain)}"
      self.profile_id = random_profile_id(self.email)
      self.gender = random.choice(gender)
      self.formerly_incarcerated = percent_true(85)

      if self.formerly_incarcerated == True:
        self.convictions_list = random.sample(convictions_list, k=randint(1, 4))
      else:
        self.convictions_list = "None"

      address = choice(addresses)
      self.street_address = address['street']
      self.city = address['city']
      self.state = address['state']
      self.zip = address['zip']
      self.veteran_status = percent_true(5)

      self.intention = random_intention()
      if self.intention == "Programming Readiness":
        self.tech_stack = random.choice(mentee_stack)
        self.career_development = "None"
      elif self.intention == "Life Coaching":
        self.tech_stack = "None"
        self.known_languages = "None"
        self.career_development = "None"
      else:
        self.tech_stack = "None"
        self.known_languages = "None"
        self.career_development = random.sample(career_development,
                                                k=randint(1, 3))

      self.known_languages = random_mentee_langauge()
      self.date_submitted = random_date(2020, 1, 1, 2022, 1, 2)
  
  def __str__(self):
      return "\n".join(f"{k}: {v}" for k, v in vars(self).items()) + "\n"

  @classmethod
  def increment_intake_counter(cls):
    cls.mentee_intake_id += 1
    return cls.mentee_intake_id

  @classmethod
  def to_df(cls, num_rows):
      return pd.DataFrame(vars(cls()) for _ in range(num_rows))

mentees = Mentee.to_df(1000)
mentees
```
