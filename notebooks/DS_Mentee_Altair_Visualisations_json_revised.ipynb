```python
import requests
import pandas as pd
import altair as alt
import json
```

```python
requests.get("http://underdog-devs-ds-a-dev.us-east-1.elasticbeanstalk.com/version").json()
```

```python
mentees_df = pd.DataFrame(requests.post("http://underdog-devs-ds-a-dev.us-east-1.elasticbeanstalk.com/Mentees/read").json()["result"])
```

```python
MenteesChart = alt.Chart(mentees_df).mark_bar().encode(
    x='experience_level',
    y='count()',
    color='experience_level',
).properties(
    title="Mentees: Experience Level",
    width=500,
    height=600,
)
```

```python
MenteesChart # remove nulls?
```

```python
dirty_json = MenteesChart.to_json()
```

```python
dirty_json # Not friendly to web due to the line feeds added to make it print pretty in Python
```

```python
clean_json_object = json.loads(dirty_json)
```

```python
clean_json_object # Now it's a clean JSON object :) This is what we can try to send to backend to be used with the vegaEmbed library for frontend
```

```python
# This saves the JSON object locally as a .json file
# Not necessary for API implementation, but a potential way to send backend an example JSON object.
with open('MenteesChart.json', 'w') as f:
    json.dump(clean_json_object, f)
```
