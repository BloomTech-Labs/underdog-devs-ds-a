```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def clean_data(df):
    columns_needed = ['skills','skill_rank','time_zone','tech_or_career']
    df = df[columns_needed]
    df = pd.concat([pd.get_dummies(df[c].apply(pd.Series).stack()).sum(level=0) for c in columns_needed], axis=1)
    return df
```
