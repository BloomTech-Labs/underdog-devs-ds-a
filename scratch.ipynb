```python
import requests
import altair as alt
import pandas as pd

from app.data import MongoDB
```

```python
def tech_stack_by_role(dataframe: pd.DataFrame):
    return alt.Chart(
        dataframe,
        title='Tech Stack by Role',
    ).mark_bar(opacity=0.7).encode(
        x=alt.X('count()', title="", stack=None),
        y=alt.Y('tech_stack', title=""),
        color=alt.Color('user_role', title="User Role"),
        # tooltip=alt.Tooltip(list(dataframe.columns)),
    )
```

```python
db = MongoDB("UnderdogDevs")
mentors_df = pd.DataFrame(db.read("Mentors"))[["tech_stack", "name"]]
mentees_df = pd.DataFrame(db.read("Mentees"))[["tech_stack", "name"]]
mentors_df["user_role"] = "Mentor"
mentees_df["user_role"] = "Mentee"
df = pd.concat([mentees_df, mentors_df], axis=0).reset_index(drop=True)
df
```

```python
tech_stack_by_role(df)
```
